```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from numpy.linalg import eig
```

```python
x = [4, 8, 13, 7]
y = [11, 4, 5, 14]
```

```python
xMean = np.mean(x)
yMean = np.mean(y)
print(f"Mean of x = {xMean}\nMean of y = {yMean}")
```

```python
plt.scatter(x, y, color="black", marker="v")
plt.scatter(xMean, yMean, label="Mean", color="red", marker="*")
plt.annotate("Mean", (xMean, yMean))
plt.title("Initital Plot with given Dataset")
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
plt.grid()
plt.xlim(-15, 15)
plt.ylim(-15, 15)
plt.show()
```

```python
xCentered = np.array(x - xMean)
yCentered = np.array(y - yMean)
xCenteredMean = np.mean(xCentered)
yCenteredMean = np.mean(yCentered)

plt.title("Plot with Mean Normalized Data")
plt.scatter(xCentered, yCentered, color="black", marker="v")
plt.scatter(xCenteredMean, yCenteredMean, label="Mean", color="red", marker="*")
plt.annotate("Mean", (xCenteredMean, yCenteredMean))
plt.xlim(-15, 15)
plt.ylim(-15, 15)
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
plt.grid()
plt.show()
```

```python
covarianceMatrix = np.cov(np.array((x, y)))
print(f"Covariance matrix\n{covarianceMatrix}")
```

```python
eigenValues, eigenVectors = eig(covarianceMatrix)
idx = np.argsort(eigenValues)[::-1]
eigenValues = eigenValues[idx]
eigenVectors = eigenVectors[:, idx]
print(f"Eigen values = {eigenValues}\nEigen vectors\n{eigenVectors}")

plt.title("Plot of EigenVectors")
plt.scatter(xCentered, yCentered, color="black", marker="v")
plt.scatter(xCenteredMean, yCenteredMean, label="Mean", color="red", marker="*")
plt.xlim(-15, 15)
plt.ylim(-15, 15)
plt.quiver(0, 0, eigenVectors[0][0], eigenVectors[0][1], scale=(100/30.3), color='red')
plt.quiver(0, 0, eigenVectors[1][0], eigenVectors[1][1], scale=(100/6.6), color='blue')
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
plt.grid()
plt.show()
```

```python
def getFoot(p, a, b):
    ap = p - a
    ab = b - a
    result = a + np.dot(ap, ab)/np.dot(ab, ab) * ab
    return result

foot = getFoot((xCentered[0], yCentered[0]), np.array([0, 0]), (eigenVectors[0][0], eigenVectors[0][1]))

plt.title("Transformation of 2D data to 1D data")
plt.scatter(xCentered, yCentered, color="black", marker="v")
plt.xlim(-15, 15)
plt.ylim(-15, 15)
plt.axline((0, 0), (5, -10), color="purple")
plt.quiver(0, 0, eigenVectors[0][0], eigenVectors[0][1], scale=(100/30.3), color='r')
for i in range(4):
    foot = getFoot((xCentered[i], yCentered[i]), np.array([0, 0]), (5, -10))
    plt.plot(foot[0], foot[1], marker="P", color="brown")
plt.grid()
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
plt.show()
```

```python
k = 1
topK = eigenVectors[:, :k]
transformedData = np.dot(topK.T, np.array((xCentered, yCentered)))

plt.scatter(transformedData, np.zeros_like(transformedData), marker='P', color='brown')
plt.title("Orientation and Comparison with original data")
plt.scatter(x, y, color="black", marker="v")
plt.scatter(xMean, yMean, label="Mean", color="red", marker="*")
plt.annotate("Mean", (xMean, yMean))
plt.axline((np.amin(transformedData), 0), (np.amax(transformedData), 0), color="purple")
plt.grid()
plt.xlim(-15, 15)
plt.ylim(-1, 15)
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
plt.show()
```
