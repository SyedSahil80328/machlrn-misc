```python
# importing the required module
import matplotlib.pyplot as plt
x = [4,8,13,7]
dup_x = x.copy()
y = [11,4,5,14]
dup_y = y.copy()
plt.scatter(x, y)


plt.xlabel('x - axis')

plt.ylabel('y - axis')

plt.title('Initial Plotting')

plt.show()
```

```python
sum_x=0
sum_y=0
for i in range(len(x)):
    sum_x += x[i]
    sum_y += y[i]
    
mean_x = sum_x / len(x)
mean_y = sum_y / len(y)
print("Mean of X : ",mean_x)
print("Mean of Y : ",mean_y)                                                

for i in range(len(x)):
    x[i] = x[i]-mean_x
    y[i] = y[i]-mean_y

print(x)
print(y)
```

```python
plt.scatter(x, y)

plt.xlabel('x - axis')

plt.ylabel('y - axis')

plt.title('After shifting towards Center Plotting')

plt.show()
```

```python
import statistics
import numpy as np
var_x = statistics.variance(dup_x)
var_y = statistics.variance(dup_y)
print("Variance of X : ",var_x)
print("Variance of Y : ",var_y)
cov_mat = np.stack((x, y), axis = 0) 
print("Covariance Matrix")
covariance_matrix=np.cov(cov_mat)
print(covariance_matrix)
```

```python
w, v = np.linalg.eig(covariance_matrix) 
  
print("Printing the Eigen values of the given square array:\n", w) 

print("Printing Right eigenvectors of the given square array:\n",v)
```

```python
w[::-1].sort()
principal_component=w[0]
transformed_data = np.dot(np.array([x, y]).T, principal_component.T)
plt.scatter(transformed_data, np.zeros_like(transformed_data))
plt.xlabel('Principal Component')
plt.title('Transformed Data (1D)')
plt.grid()
plt.show()
```

```python
print("Transformed Data (1D):")
for d in transformed_data:
    print(d)
```
